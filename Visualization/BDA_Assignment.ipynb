```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np

df = pd.read_csv('Time_records.csv')
df.head()
```

**1.Year wise carrier delay from 2003-2010**

SQL:SELECT Year, avg((CarrierDelay /ArrDelay)*100) from flight_delays GROUP BY Year 

```python
x_pos = np.arange(len(df['Iteration']))
width = 0.2



fig, ax = plt.subplots(figsize=(6, 4), dpi=300)
ax.bar(x_pos, df['SQL1_Hive'], width=width, align='center', label='Hive')
ax.bar(x_pos+width, df['SQL1_Spark'], width=width, align='center', label='Spark')


labels = df['Iteration']
ax.set_xticks(x_pos + width)
ax.set_xticklabels(labels, size=8)

ax.bar_label(ax.containers[0], size=5)
ax.bar_label(ax.containers[1], size=5)


ax.set_xlabel("Iteration")
ax.set_ylabel("Runnig Time (s)")
ax.set_title("Year wise carrier delay from 2003-2010")

ax.legend()


```

Calculating average of time taken for each sql by hadoop and spark

```python
av_column = df.mean(axis=0)
av_column.to_frame()
```

```python
data = {'Query': ['carrier delay query', 'NAS delay query', 'Weather delay query', 'late aircraft delay query','security delay query'],
        'Hive': [av_column['SQL1_Hive'], av_column['SQL2_Hive'],av_column['SQL3_Hive'], av_column['SQL4_Hive'],av_column['SQL5_Hive']],
        'Spark': [av_column['SQL1_Spark'], av_column['SQL2_Spark'],av_column['SQL3_Spark'], av_column['SQL4_Spark'],av_column['SQL5_Spark']]}

df2 = pd.DataFrame(data)
df2
```

```python
x_pos = np.arange(len(df2['Query']))
width = 0.2



fig, ax = plt.subplots(figsize=(6,10), dpi=300)
ax.bar(x_pos, df2['Hive'], width=width, align='center', label='Hive')
ax.bar(x_pos+width, df2['Spark'], width=width, align='center', label='Spark')


labels = df2['Query']
ax.set_xticks(x_pos + width)
ax.set_xticklabels(labels, size=5)

ax.bar_label(ax.containers[0], size=4)
ax.bar_label(ax.containers[1], size=4)

ax.set_xlabel("Query")
ax.set_ylabel("Time Taken avg (s)")
ax.set_title("Time Taken By")

ax.legend()
```
